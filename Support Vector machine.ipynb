```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



```

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)

```

```python
y.shape
```

```python
y
```

```python
y_new=np.where(1> y,y,1)
   
```

```python
y_new
```

```python
plt.plot(X[:,0][y_new == 0 ],X[:,1][y_new == 0 ],'b.',label= 'setosa')
plt.plot(X[:,0][y_new == 1 ],X[:,1][y_new == 1 ],'r.',label= 'Versicolour/verginia')

plt.legend()
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.svm import SVC
svm_linear = SVC(kernel='linear', decision_function_shape='ovr')
svm_linear.fit(X, y_new)
```

```python
print(svm_linear.support_vectors_)
```

```python
from sklearn.svm import SVC
svm_rbf = SVC(kernel='rbf', decision_function_shape='ovr')
svm_rbf.fit(X, y_new)
```

```python

print(svm_rbf.support_vectors_)
```

```python
from sklearn.svm import SVC
svm_poly = SVC(kernel='poly', decision_function_shape='ovr')
svm_poly.fit(X, y_new)
```

```python
print(svm_poly.support_vectors_)
```

```python
w1 = svm_linear.coef_[0,0]
w2 = svm_linear.coef_[0,1]
b = svm_linear.intercept_[0]
print("Parameters learned for the SVM classifier: ")
print("w1 = {:.4f}, w2 = {:.4f}, b = {:.4f}".format(w1, w2, b))
print("\nEquation of the decision boundary: ")
print("f(x)={:.4f}*x1 + {:.4f}*x2 + ({:.4f}) = 0".format(w1, w2, b))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_new, test_size=0.25)
```

```python
svm_linear_accuracy = svm_linear.score(X_train, y_train)
print("Training set accuracy of linear SVM =", svm_linear_accuracy)

svm_poly_accuracy = svm_poly.score(X_train, y_train)
print("Training set accuracy of poly SVM =", svm_poly_accuracy)

svm_rbf_accuracy = svm_rbf.score(X_train, y_train)
print("Training set accuracy of rbf SVM =", svm_rbf_accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

fig, ax = plt.subplots( figsize=(13,4.5))
pred=svm_linear.predict(X_test)

cm_test = confusion_matrix(y_test, pred, labels=svm_linear.classes_)
ConfusionMatrixDisplay(confusion_matrix=cm_test, 
                     display_labels= ['0', '1']).plot(ax=ax, cmap='Purples')
ax.set_title('Test confusion matrix ')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

fig, ax = plt.subplots( figsize=(13,4.5))
pred=svm_rbf.predict(X_test)

cm_test = confusion_matrix(y_test, pred, labels=svm_rbf.classes_)
ConfusionMatrixDisplay(confusion_matrix=cm_test, 
                     display_labels= ['0', '1']).plot(ax=ax, cmap='Reds')
ax.set_title('Test confusion matrix ')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

fig, ax = plt.subplots( figsize=(13,4.5))
pred=svm_poly.predict(X_test)

cm_test = confusion_matrix(y_test, pred, labels=svm_poly.classes_)
ConfusionMatrixDisplay(confusion_matrix=cm_test, 
                     display_labels= ['0', '1']).plot(ax=ax, cmap='Greens')
ax.set_title('Test confusion matrix ')
plt.show()
```
